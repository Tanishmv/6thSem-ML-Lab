**Canada per capita income**

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample_data/canada_per_capita_income.csv')
df.head()
```

```python
print(df.isnull().sum())
```

```python
plt.xlabel('year')
plt.ylabel('per_captia_income')
plt.scatter(df.year, df["per capita income (US$)"], color='red', marker='+')

new_df = df.drop('per capita income (US$)',axis='columns')
new_df.head()

```

```python
per_capita_income=df["per capita income (US$)"]
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income)
```

```python
reg.predict([[2020]])

```

```python
reg.coef_
reg.intercept_
```

**Hiring**

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df2 = pd.read_csv('/content/sample_data/hiring.csv')
df2.head()
```

```python
print(df2.isnull().sum())
```

```python

word_to_number = {
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11,
    'twelve':12,
    'thirteen':13,
    'zero':0,
}

df2.experience = df2.experience.map(word_to_number)

median_experience = df2.experience.median()

df2.experience = df2.experience.fillna(median_experience)

df2.head()
```

```python
df2["test_score(out of 10)"].median()
df2["test_score(out of 10)"] = df2["test_score(out of 10)"].fillna(df2["test_score(out of 10)"].median())
df2
```

```python
reg2 = linear_model.LinearRegression()
reg2.fit(df2.drop('salary($)',axis='columns'),df2['salary($)'])
```

```python
reg2.predict([[2, 9, 6]])
```

```python
reg2.predict([[12, 10, 10]])
```

**1000 Companies**

```python
df3 = pd.read_csv('/content/sample_data/1000_Companies.csv')
df3.head()
```

```python
print(df3.isnull().sum())
```

```python
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for column in df3.columns:
    if df3[column].dtype == 'object' and column != 'Profit':
        df3[column] = encoder.fit_transform(df3[column])
```

```python
df3.head()
```

```python
reg3 = linear_model.LinearRegression()
reg3.fit(df3.drop('Profit',axis='columns'),df3.Profit)
```

```python
reg3.predict([[91694.48, 515841.3, 11931.24, 1]])
```
